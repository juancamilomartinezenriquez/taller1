```python
# -*- coding: utf-8 -*-
"""Untitled0.ipynb

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/12lUT4Mi2I5pZ-i41lWIRIMnJi_Nvws01
"""

!pip install scikit-fuzzy

import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# Variables de entrada
cantidad_dinero = ctrl.Antecedent(np.arange(0,501,1),'cantidad_dinero')
salario_cliente = ctrl.Antecedent(np.arange(0,21,1),'salario_cliente')
# Variables de salida
probabilidad_preferencial = ctrl.Consequent(np.arange(0,101,1),'probabilidad_preferencial')

## Funciones de membresia

## Cantidad de dinero

cantidad_dinero['muy baja']=fuzz.trimf(cantidad_dinero.universe,[0,0,20])
cantidad_dinero['baja']=fuzz.trimf(cantidad_dinero.universe,[20,30,40])
cantidad_dinero['media']=fuzz.trapmf(cantidad_dinero.universe,[30,40,50,60])
cantidad_dinero['alta']=fuzz.trapmf(cantidad_dinero.universe,[50,60,500,500])

cantidad_dinero.view()

## Salario del cliente

salario_cliente['muy bajo']=fuzz.trimf(salario_cliente.universe,[0,0,2])
salario_cliente['bajo']=fuzz.trimf(salario_cliente.universe,[2,3,4])
salario_cliente['medio']=fuzz.trimf(salario_cliente.universe,[3,6,8])
salario_cliente['alto']=fuzz.trapmf(salario_cliente.universe,[7,8,21,21])

salario_cliente.view()

## Probabilidad

probabilidad_preferencial['muy baja']=fuzz.trimf(probabilidad_preferencial.universe,[0,0,20])
probabilidad_preferencial['baja']=fuzz.trimf(probabilidad_preferencial.universe,[15,40,60])
probabilidad_preferencial['media']=fuzz.trimf(probabilidad_preferencial.universe,[50,70,90])
probabilidad_preferencial['alta']=fuzz.trimf(probabilidad_preferencial.universe,[80,100,100])

probabilidad_preferencial.view()

## Reglas difusas

regla11 = ctrl.Rule(cantidad_dinero['muy baja']&salario_cliente['muy bajo'],probabilidad_preferencial['muy baja'])
regla12 = ctrl.Rule(cantidad_dinero['baja']&salario_cliente['muy bajo'],probabilidad_preferencial['muy baja'])
regla13 = ctrl.Rule(cantidad_dinero['media']&salario_cliente['muy bajo'],probabilidad_preferencial['baja'])
regla14 = ctrl.Rule(cantidad_dinero['alta']&salario_cliente['muy bajo'],probabilidad_preferencial['media'])

regla21 = ctrl.Rule(cantidad_dinero['muy baja']&salario_cliente['bajo'],probabilidad_preferencial['muy baja'])
regla22 = ctrl.Rule(cantidad_dinero['baja']&salario_cliente['bajo'],probabilidad_preferencial['baja'])
regla23 = ctrl.Rule(cantidad_dinero['media']&salario_cliente['bajo'],probabilidad_preferencial['media'])
regla24 = ctrl.Rule(cantidad_dinero['alta']&salario_cliente['bajo'],probabilidad_preferencial['media'])

regla31 = ctrl.Rule(cantidad_dinero['muy baja']&salario_cliente['medio'],probabilidad_preferencial['baja'])
regla32 = ctrl.Rule(cantidad_dinero['baja']&salario_cliente['medio'],probabilidad_preferencial['baja'])
regla33 = ctrl.Rule(cantidad_dinero['media']&salario_cliente['medio'],probabilidad_preferencial['media'])
regla34 = ctrl.Rule(cantidad_dinero['alta']&salario_cliente['medio'],probabilidad_preferencial['alta'])

regla41 = ctrl.Rule(cantidad_dinero['muy baja']&salario_cliente['alto'],probabilidad_preferencial['baja'])
regla42 = ctrl.Rule(cantidad_dinero['baja']&salario_cliente['alto'],probabilidad_preferencial['media'])
regla43 = ctrl.Rule(cantidad_dinero['media']&salario_cliente['alto'],probabilidad_preferencial['alta'])
regla44 = ctrl.Rule(cantidad_dinero['alta']&salario_cliente['alto'],probabilidad_preferencial['alta'])

sistema_control = ctrl.ControlSystem(
    [
        regla11,
        regla12,
        regla13,
        regla14,
        regla21,
        regla22,
        regla23,
        regla24,
        regla31,
        regla32,
        regla33,
        regla34,
        regla41,
        regla42,
        regla43,
        regla44,
    ]
)

controlador = ctrl.ControlSystemSimulation(sistema_control)

controlador.input['cantidad_dinero']=30
controlador.input['salario_cliente']=9

controlador.compute()

probabilidad_calculada = controlador.output['probabilidad_preferencial']
print(f"Probablidad {probabilidad_calculada}")
```
